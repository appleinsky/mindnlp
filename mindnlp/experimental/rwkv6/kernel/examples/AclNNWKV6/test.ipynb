```python
q_input = "/root/wkv6/AclNNInvocationNaive/input/input_q.npy"
q_result = np.load(q_input) 
k_input = "/root/wkv6/AclNNInvocationNaive/input/input_k.npy"
k_result = np.load(k_input) 
v_input = "/root/wkv6/AclNNInvocationNaive/input/input_v.npy"
v_result = np.load(v_input) 
w_input = "/root/wkv6/AclNNInvocationNaive/input/input_w.npy"
w_result = np.load(w_input) 
u_input = "/root/wkv6/AclNNInvocationNaive/input/input_u.npy"
u_result = np.load(u_input) 
h_input = "/root/wkv6/AclNNInvocationNaive/input/input_h0.npy"
h_result = np.load(h_input) 
```

```python
B = 1
T = 64
N = 64
H = 1
param_shape = (B, H, T, N)
state_shape = (B, H, N, N)
h = h_result
o = np.zeros(param_shape)
```

```python
for b_i in range(B):
        for h_i in range(H):
            for i in range(N):
                for t in range(T):
                    for j in range(N):
                        x = (k_result[b_i, h_i, t, j] * 0.5) * (v_result[b_i, h_i, t, i] * 0.5)
                        s = h[b_i, h_i, j, i]
                        # print(h)
                        o[b_i, h_i, t, i] += q_result[b_i, h_i, t, j] * (u_result[h_i, j] * x + s)
                        # we are actually using exp(-exo(w)) as decay coefficient, which is always within (0,1)
                        h[b_i, h_i, j, i] = s * np.exp(-np.exp(w_result[b_i, h_i, t, j])) + x
            print(o)
                        
```
